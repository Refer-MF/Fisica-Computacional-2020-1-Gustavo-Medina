$A=\pi r^2$



```
#Codigo para calcular el area de un circulo simple
print("CALCULANDO EL AREA DE UN CIRCULO CON RADIO r")
def ac(x):
  return np.pi*x**2 #Definiendo la formula para calcular el area de un circulo.

#===============INTERACCION CON EL USUARIO==========================
print("Por favor ingrese el valor del radio del circulo en cuestion:")
r=float(input()) #Radio del circulo a calcular

ac=ac(r)


print("\nArea del circulo con radio", r, ":", ac, "u^2") #Impresion del resultado
print("\nArea del circulo con dos puntos decimales = {:.2f}".format(ac), "u^2")
```

## Formula usada:

```
#Codigo para calcular el volumen de una esfera.
print("CALCULANDO EL VOLUMEN DE UNA ESFERA")
def Ves(x):
  return 4/3*np.pi*x**3 #Definiendo la formula para calcular el volumen de una esfera

#===============INTERACCION CON EL USUARIO==========================
print("Por favor ingrese el valor del radio de la esfera en cuestion:")
r=float(input()) #Radio de la esfera en cuestion

Ves=Ves(r)


print("\nVolumen de la esfera:", Ves, "u^3") #Impresion del resultado
print("\nVolumen de la esfera con dos puntos decimales: {:.2f}".format(Ves), "u^3")
```

# $\underline{\text{Volumen de un cilindro circular}}$

### Formula usada:

$V= (\text{Area del circulo de la base})(h)$ 

o

$V= (\pi r^2)(h)$

```
#Codigo para calcular el volumen de un cilindro circular.
print("CALCULANDO EL VOLUMEN DE UN CILINDRO CIRCULAR")
#Para esto necesitaremos el area del circulo en la base del cilindro

def Vci(x,y):
  return (np.pi*x**2)*y #Definiendo la formula para calcular el area de un circulo y el volumen del cilindro

#===============INTERACCION CON EL USUARIO==========================
print("Por favor ingrese el valor del radio del cilindro en cuestion:")
r=float(input()) #Radio del cilindro en cuestion

print("Ahora ingrese el valor de la altura del cilindro:")
h=float(input()) #Altura del cilindro

Vci=Vci(r,h)


print("\nVolumen del cilindro circular:", Vci, "u^3") #Impresion del resultado
print("\nVolumen del cilindro circular con dos puntos decimales: {:.2f}".format(Vci), "u^3")
```

# $\underline{\text{Ejercicio 2}} \text{(Codigo de raizes)}$

#Formula cuadratica:

$\frac{-b \pm \sqrt{b^2-4ac}}{2a}$

```
#a, b y c seran los coeficientes de la funcion cuadratica.

#Importanto una libreria para poder calcular raices complejas
import cmath 


#===BASURA=============
##def sqrt(x):
 ## return x**(1/2)
#===BASURA=============



#Definicion de la formula de la raiz positiva
def pos(x,y,z):
  return (-y+cmath.sqrt(z))/(2*x)

#Definicion de la formula de la raiz negativa
def neg(x,y,z):
  return (-y-cmath.sqrt(z))/(2*x)
  

#================INTERACCION CON EL USUARIO=======================
print("CALCULANDO LAS RAICES DE UNA FUNCION CUADRATICA")
print("\nIntroduzca los coeficientes de su funcion cuadratica; primero 'a', luego 'b' y por ultimo 'c':")

a=float(input())
b=float(input())
c=float(input())

f=b**2-(4*a*c)


#===BASURA=============
##g=sqrt(f)
#===BASURA=============

rapos=pos(a,b,f) #Raiz positiva
raneg=neg(a,b,f) #Raiz negativa

print("\nLa raiz x1 de la funcion es:", rapos)
print("\nLa raiz x2 de la funcion es:", raneg)

```

# Formula usada:

$x_{n+1}=\frac{1}{2}(x_n+\frac{S}{x_n})$, donde $x_{n+1}$ es la media entre $S$ y $x_n$


```

#Importando numpy
import numpy as np


#=================INTERACCION CON EL USUARIO=======================
#Pidiendo el numero S al usuario:

print("Hola. La funcion de este programa es la de aproximar la raiz de un numero arbitrario con el metodo babiliano (o de Heron).")
print("\nAhora, por favor ingrese el numero al cual querra aproximar su raiz:")

S=float(input())
Srai=np.sqrt(S) #La raiz "exacta" del numero S, calculado por numpy

print("\nBien, ahora deme un valor inicial de x0:")

x0=float(input())

#============METODO=============
i=1
aprox=1/2*(x0+(S/x0))

#==========LOOP=================
while abs(S-aprox) > 0.01:
  print("Ciclo", i, ":", aprox)
  aprox=1/2*(aprox+(S/aprox))
  i=i+1
  if i >= 10000: #Esto evitara que el loop corra infinitamente y corra a lo maximo 10,000 veces (se considero suficiente)
    break

print("El valor aproximado de la raiz de ", S, "es:", aprox, ". \nY su error es de:", abs(np.sqrt(S)-aprox))
print("\nGracias por usar este programa!")
print("\n-Por Gustavo Medina.")
```
