```
#Importando las bibliotecas que usaremos:

import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
from datetime import datetime
from math import exp, sqrt
```

```
#Ahora vamos a definir la funcion del metodo de Euler para ED:
def odeEuler(f,y0,t):
    '''Aproxima la solución de y'=f(y,t) por el método de Euler.
    
    Parámetros
    ----------
    f : La función dada por el lado derecho de la ecuación diferencial 
         y'=f(t,y), y(t_0)=y_0
    y0 : El valor inicial de y al tiempo y(t0)=y0 
    t : Arreglo unidimensional de 1D. El tamaño del paso h = t[n+1] - t[n].
    
    Regresa
    -------
    y : 1D Arreglo con la aproximación de y[n] de la solución y(t_n) calculada 
    con el método de  Euler.
    '''
    # Definimos el arreglo para las soluciones del tamaño del tiempo de integración
    y = np.zeros(len(t))
    
    # La condición inicial que debe satisfacer la solución
    y[0] = y0
    
    # LA integración por el método de Euler
    for n in range(0,len(t)-1):
        y[n+1] = y[n] + f(y[n],t[n])*(t[n+1] - t[n])
    return y
```

# $\underline{\text{Ejercicio 1:}}$


Ejercicio 1:

Resuelva la ecuación diferencial del oscilador de Van der Pol
$$\begin{equation*} \frac{d^2 x}{d t^2} - \mu (1 - x^2) \frac{dx}{dt } + x = 0 \end{equation*}$$

donde $x$ es la posición y $\mu$ es un parámetro de la parte no lineal.

Resuelva el caso para $\mu = 0, 1, 2, 3, 4$ y grafique las soluciones para un tiempo de integración de t = [0, 50]


```
#Ahora definimos la funcion para resolver ED de segundo orden:

def OsciladorVanderPol(y,t,mu):
        x, v = y
        dydt = [v, mu*(1 - x**2)*v - x]
        return dydt

#Condiciones iniciales x(0)=1, v(0)=0
y0 = [1.0,0.0]

# Definimos el tiempo de integración de 0 a 50 segundos 
t = np.linspace(0, 50, 1000)

abserr = 1.0e-8
relerr = 1.0e-6

#Para mu=0
mu1 = [0.0,1.0,2.0,3.0,4.0]

for mu in mu1:

# Usamos el método de integración de scipy.integrate.odeint
  y = odeint(OsciladorVanderPol, y0, t, args=(mu,), atol=abserr, rtol=relerr)



  plt.figure(figsize=(10,5))
  plt.plot(t, y[:, 0], 'darkorange', label=('x(t) de mu=',mu))
  plt.plot(t, y[:, 1], 'darkblue', label=('v(t) de mu=',mu))
  plt.legend(loc='best')
  plt.title('Solución de la ecuación de movimiento del Oscilador de Van der Pol')
  plt.xlabel('t')
  plt.grid()
  plt.show()
```

Ejercicio 3.1



```
def Ej3(y,t,k,g):
        x, v = y
        dydt = [v, -k*v**2 +g]
        return dydt

k = 1.0
g = 9.81
x_0 = 0.0
v_0 = 0.0

# Definimos el tiempo de integración de 0 a 50 segundos 
t = np.linspace(0, 10, 100)
```

```


#Solución con método de Euler
f = lambda v,t: -k*v**2 + g

#Para f1t
v = np.zeros(len(t))
v[0] = v_0
    
for n in range(0,len(t)-1):
  v[n+1] = v[n] + f(v[n],t[n])*(t[n+1] - t[n])


#Para f2t
x = np.zeros(len(t))
x[0] = x_0
    
for n in range(0,len(t)-1):
  x[n+1] = x[n] + v[n]*(t[n+1] - t[n])


plt.figure(figsize=(10,5))
plt.plot(t,v,'g-', t,x,'b-')
plt.legend(['v(t)', 'x(t)'])
plt.axis([0,2,0,9])
plt.grid()
plt.title('Solución con método de Euler')
plt.xlabel('t')
plt.show()

```

```
def runge_kutta(f, x_0, y_0, h):
    """Four step Runge-Kutta method (RK4)
    Solves first order ODEs
    """
    k_0 = f(x_0, y_0)
    k_1 = f(x_0 + h/2, y_0 + h/2 * k_0)
    k_2 = f(x_0 + h/2, y_0 + h/2 * k_1)
    k_3 = f(x_0 + h, y_0 + h * k_2)

    k = 1/6 * (k_0 + 2.0*k_1 + 2.0*k_2 + k_3)

    x_1 = x_0 + h
    y_1 = y_0 + h * k

    return x_1, y_1

```

```

# Condiciones iniciales
t_0 = 0.0

k = 1
g = 9.81

v1=0
t=0

# Paso de integración h = (b-a)/Npts
a=0.0
b=10.0
Npts=100
h = (b-a)/Npts




f = lambda v,t: -k*v1**2 + g
t_values = [t_0]
v_values = [v_0]

for _ in range(Npts):
  t, v1 = runge_kutta(f, t, v1, h)
  t_values.append(t)
  v_values.append(v1)


#Para la segunda ecuación 
#Solve_ivp


# Definimos la parte derecha de la ecuación

def P2(t,y,k,g):
        dydt = [y[1], -k*(y[1])**2 +g]
        return dydt


# Condiciones iniciales
y_0 = [x_0, v_0]

t = np.linspace(t_0,b,Npts)

abserr = 1.0e-8
relerr = 1.0e-6


sol = solve_ivp(lambda t, y, mu: P2(t, y, k, g), (t_0,b), y0=y_0, 
                method='RK45', t_eval=t, args= (mu,), rtol=relerr, atol=abserr)

xp2 = sol.y[0]



# Graficar la solución
plt.figure(figsize=(10,5))
#primera, calculada con RK4
plt.plot(t_values, v_values, 'g', label='v(t)')
#segunda, calculada con solve_ivp
plt.plot(t, xp2, 'b', label='x(t)')
plt.axis([0,2,0,9])
plt.legend(loc='best')
plt.title('Solución de la EDO usando Runge-Kutta RK4')
plt.xlabel('t')
plt.grid()
plt.show()
```

```
#Solución con Odeint
y = odeint(Ej3, y_0, t, args=(k,g), atol=abserr, rtol=relerr)

plt.figure(figsize=(10,5))
plt.plot(t, y[:,0], 'b', label='x(t)')
plt.plot(t, y[:,1], 'g', label='v(t)')
plt.axis([0,2,0,9])
plt.legend(loc='best')
plt.title('Solución con odeint')
plt.xlabel('t')
plt.grid()
plt.show()

```
